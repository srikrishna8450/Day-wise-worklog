### write a function to check
- spcified-0
- low-1
- medium-2
- high-3
- critical-4


```python
def cat_to_num(x):
    if x=='NOT SPCIFIED':
        return 0
    elif x=='LOW':
        return 1
    elif x=='MEDIUM':
        return 2
    elif x=="HIGH":
        return 3
    elif x=='CRITICAL':
        return 4
    
```

```python
data["Order_Priority_Number"]=data['Order_Priority'].apply(cat_to_num)
data["Order_Priority_Number"].value_counts()
```
