```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_regression
```

```python
help(make_regression)
```

```python
X,y =make_regression(100,1,n_targets=1,random_state =34)
X
```

```python
y
```

```python
print(X.shape)
print(type(X))
print(X.ndim)
print(y.shape)
print(type(y))
print(y.ndim)
```

```python
plt.plot(X,y,"or")
```

```python
# import the model linear regressor model
from sklearn.linear_model import LinearRegression
# create an instance
LR = LinearRegression()
# train
LR.fit(X,y)
# predict with train data
y_pred =LR.predict(X)
```

```python
print(LR.coef_)
print(LR.intercept_)
```

```python
# create user defined function with name mse
# inputs y_pred, y
# output is mean squared error
```

```python
def mse(y_pred,y):
    return ((y_pred - y)**2).mean()
    
```

```python
# create user defined function with name mse
# inputs y_pred, y
# output is mean squared error
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_pred,y))
```
